```julia
function rk4(x::Vector,u,dt::Float64, dynamics::Function)
    # vanilla RK4
    k1 = dt*dynamics(x, u)
    k2 = dt*dynamics(x + k1/2, u)
    k3 = dt*dynamics(x + k2/2, u)
    k4 = dt*dynamics(x + k3, u)
    return x + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
end
```

```julia
function diff_drive_dynamics(x,u)
    
#     @show x, u
    
    Vx = u[1]
    Vy = u[2]
    ω = u[3]
    
    θ = x[3]
    
#     @show θ
    
    
    ẋ = [Vx * cos(θ),
        Vy * sin(θ),
        ω]
    
    return ẋ
end
```

```julia
function ϕ(x)
    p = x[1:2]  
    -(norm(p - [3,3])^2 - 1.5^2) 
end
function ∇ϕ(x)
    ForwardDiff.gradient(ϕ, x)
end
function cbf(x::Vector,  # current state 
             u0::Vector, # nominal control input 
             A::Matrix,  # discrete time A 
             B::Matrix,  # discrete time B 
             dt::Real)   # time step size 
    
    # tune this (lower the number = more aggressive)
    λ = -5
    
    # our variable we are solving for 
    u = cvx.Variable(2)
    
    # let's approximate xdot with a difference between (x_{k+1} - x_k)/dt
    # this is because if we just did the continuous time dynamics, ϕdot would 
    # be zero. This is because if x = [p;v], dϕ/dx = [dϕ/dp; 0] since it only depends on 
    # the position, not velocity. our continous time dynamics are xdot = [v;u], meaning when
    # we take ∇ϕ'ẋ, we get [dϕ/dp; 0]'[v;u], which is not a function of u. This is because u
    # cannot directly affect ṗ. If instead we average xdot over a time step, u can now affect
    # ṗ, and everything works out. 
    xdot = ((A*x + B*u) - x)/dt
    gradient = ∇ϕ(x)
    ϕdot = transpose(gradient)*xdot 

    # find the closest control to the given u0
    prob = cvx.minimize(cvx.sumsquares(u - u0))
    
    # must be safe
    prob.constraints += [ϕdot <= λ*ϕ(x)]
    
    cvx.solve!(prob, ECOS.Optimizer; silent_solver = true)
    
    return vec(u.value)
end
function circle()
    center = [3,3]
    rad = 1.5 
    θs = range(0,2*pi,length = 100)
    points = [(rad*[cos(θ),sin(θ)]+center) for θ in θs]
    x = [p[1] for p in points]
    y = [p[2] for p in points]
    x, y
end
    
let 
    
    # robot dynamics 
    dt = 0.1
    tf = 70.0 
    t_vec = 0:dt:tf 
    N = length(t_vec)
    
    # set the goal states
    xic = [2.,25.,0.] # ic
    xg = [75., 25., 0.] # goal
    
    ## apply the reference
    ref = range(xic, xg, length=N)
    Xref = zeros(3, N)
    for i=1:N 
        x = ref[i][1]
        y = ref[i][2]
#         @show Xref[i], [x, y,tan(y/x)]
        Xref[:, i] = [x, y,0]
    end

    # sim 
    
    # set the p & d matrices
    Kp =[100 0 0;
        0 100 0;
        0 0 100]
    Kd =[1 0 0;
        0 1 0;
        0 0 1]
#     Kd = 0.1*Kd
#     Kp = 0.001*Kp
    
    # init the p and d errors
    e_p = [0;0;0]
    e_d = [0;0;0]
    
    X = zeros(3, N)
    X[:, 1] = [2;20;0]
    X_last = [0,0,0]
    
    U = zeros(2, N)
    for i = 1:(N-1)
        
        # the proportial error is the current state minus the reference
        e_p = X[:, i] - Xref[:, i]
        
        # take the discrete derivative
        if i == 1
            e_d = [0,0,0]
        else
            e_d = X[:, i] - X[:, i-1] / dt
#             @show X[2, i], X[1, i]
            theta_g = atan(Xref[2,i]-X[2,i], Xref[1,i]-X[1,i])
            theta_r = X[3, i]
            
            function angle_subtraction(a::Real, b::Real)
                c = b - a
                if c > π
                    c -= 2π
                elseif c < -π
                    c += 2π
                end
                return c
            end

            yaw_error = angle_subtraction(theta_r, theta_g)
            
            e_d[3] = yaw_error
            
        end
        
        # calculate the control u (vx_global, vy_global, omega)        
        u = (Kp * e_p) + (Kd * e_d)
        
        # turn the u 3x1 to a 2x1 by combining the magnitudes of vx_global, vy_global
#         U[1, i] = sqrt(u[1]^2 + u[2]^2)
#         U[2, i] = u[3]
        
        if u[1] > 1 
            u[1] = 1
        elseif u[1] < -1
            u[1] = -1
        end
        
        if u[2] > 1 
            u[2] = 1
        elseif u[2] < -1
            u[2] = -1
        end
        
        if u[3] > pi/3
            u[3] = pi/3
        elseif u[3] < -pi/3
            u[3] = -pi/3
        end
        
        X[:, i+1] = rk4(X[:, i],u, dt, diff_drive_dynamics)
        
    end
    
    for i=1:N
        @show X[:, i] #, U[:, i]
    end
    

    
    plot(Xref[1, :], Xref[2, :], label  = "des_trajectory", xlabel = "x", ylabel = "y", title = "Trajectory")
    plot!(X[1,:],X[2,:], label  = "exec_trajectory", xlabel = "x", ylabel = "y", title = "Trajectory")
    
#     ϕs = ϕ.(X)
#     plot(t_vec,ϕs,label = "",xlabel = "time (s)", ylabel = "ϕ", title = "Safety Index ϕ(x)")
    
end
```

```julia
function diff_drive_dynamics(x, u)
    Vx = u[1]
    ω = u[2]
    θ = x[3]
    ẋ = [Vx * cos(θ), Vx * sin(θ), ω]
    return ẋ
end

function rk4(x::Vector, u, dt::Float64, dynamics::Function)
    # vanilla RK4
    k1 = dt * dynamics(x, u)
    k2 = dt * dynamics(x + k1/2, u)
    k3 = dt * dynamics(x + k2/2, u)
    k4 = dt * dynamics(x + k3, u)
    return x + (1/6) * (k1 + 2*k2 + 2*k3 + k4)
end

# robot dynamics 
function simulate_robot(xic, xg, xs, Kp, Kd, dt, tf)
    t_vec = 0:dt:tf 
    N = length(t_vec)

    ## apply the reference
    Xref = range(xs, xg, length=N)
#     Xref = hcat([ref[i][1], ref[i][2], 0] for i in 1:N)

    # init the p and d errors
    e_p = [zeros(3) for i in 1:N]
    e_d = zeros(3)

    X = zeros(3, N)
    X[:, 1] = xic

    U = zeros(2, N-1)
    for i = 1:N-1
        # the proportional error is the current state minus the reference
        e_p[i] = X[:, i] - Xref[i]

        # take the discrete derivative
        if i == 1
            e_d = zeros(3)
        else
            
            function angle_subtraction(a::Real, b::Real)
                c = b - a
                if c > π
                    c -= 2π
                elseif c < -π
                    c += 2π
                end
                return c
            end
            
            theta_g = atan(Xref[i][2]-X[2,i], Xref[i][1]-X[1,i])
            theta_r = X[3,i]
            
#             @show theta_g, theta_r

            yaw_error = angle_subtraction(theta_r, theta_g)
            
            e_p[i][3] = yaw_error
            e_d = (e_p[i] - e_p[i-1]) / dt
            
            
        end
        
#         @show e_d, , Xref[i][1]-X[1,i]        
        
        # calculate the control u (Vx, ω)  
#         @show size(Kp), size(e_p[i]), size(Kd), size(e_d)
        
        u = (Kp * e_p[i]) + (Kd * e_d) #reshape(e_d, 3, 1))
        
        U[:, i] = [sqrt(u[1]^2 + u[2]^2), u[3]]
        
        # limit control inputs
        if U[1, i][1] > 2
            U[1, i] = 2
        elseif U[1, i] < -2
            U[1, i] = -2
        end
        
        if U[2, i] > pi/3
            U[2, i] = pi/3
        elseif U[2, i] < -pi/3
            U[2, i] = -pi/3
        elseif U[2, i] < 1e-5
            U[2, i] = 0
        end
                
        X[:, i+1] = rk4(X[:, i], U[:, i], dt, diff_drive_dynamics)
    end
    
    return X, Xref, U, N, t_vec
end


            
```

```julia
# set the goal states
xic = [2., 30., 0.] # initial state

xg = [75., 25., 0.] # goal state
xs = [2., 25., 0.] # starting state


# set the P and D matrices
Kp = [1 0 0; 0 1 0; 0 0 1]
Kd = [1 0 0; 0 1 0; 0 0 1]

# simulation parameters
dt = 0.01
tf = 70.0

X, Xref, U, N, t_vec = simulate_robot(xic, xg, xs, Kp, Kd, dt, tf)

refx = [Xref[i][1] for i = 1:N]
refy = [Xref[i][2] for i = 1:N]
simx = [X[1, i] for i = 1:N]
simy = [X[2, i] for i = 1:N];


# for i = 1:N
#    @show X[:, i] 
# end
```
